```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('UTSEUS-anjuke-real-estate.csv')
```

```python
df.head()
```

```python
sns.set()
```

```python
sns.lmplot(x='surface', y='onesquaremeter', data=df, height=10)
plt.yscale('log')
plt.xscale('log')
```

```python
sns.lmplot(x='surface', y='onesquaremeter', data=df, fit_reg=False, hue='district', height=10)
plt.yscale('log')
plt.xscale('log')
```

```python
sns.boxplot(data=df)
plt.yscale('log')
```

```python
sns.boxplot(data=df[['room','bedroom']])
```

```python
sns.violinplot(x='district', y='surface', data=df)
plt.xlim(-0.5,5.5)
plt.ylim(0, 250)
```

```python
sns.swarmplot(x='district', y='surface', data=df)
plt.xlim(-0.5,5.5)
plt.ylim(0, 250)
```
