```python
df['city'] = df['city'].apply(lambda x: "Hebei" if x == "?city=Hebei" else x.title())
```

```python
df_city = df[df['city'] == 'City']
df_station = df[df['city'] != 'City']
print(len(df_city))
print(len(df_station))
```

```python
def df_to_geojson(df, properties, lat='lat', lon='long'):
    geojson = {'type':'FeatureCollection', 'features':[]}
    for _, row in df.iterrows():
        feature = {'type':'Feature',
                   'properties':{},
                   'geometry':{'type':'Point',
                               'coordinates':[]}}
        feature['geometry']['coordinates'] = [row[lon],row[lat]]
        for prop in properties:
            feature['properties'][prop] = row[prop]
        geojson['features'].append(feature)
    return geojson
```

```python
cols = ['city', 'name', 'aqi', 'pm25', 'pm10', 'no2', 'so2', 'o3', 'co', 'temp']
geojson = df_to_geojson(df_station, cols)
```

```python
import json
output_filename = 'dataset.geojson'
with open(output_filename, 'w') as handler:
    #output_file.write('const dataset = ')
    json.dump(geojson, handler) 
```
