```python
import requests as rq

top_cities = ["Shanghai","Beijing","Guangzhou","Shenzhen","Tianjin","Wuhan","Dongguan","Chengdu","Foshan","Chongqing","Nanjing","Shenyang","Hangzhou","Xi'an","Harbin","Suzhou","Qingdao","Dalian","Zhengzhou","Shantou","Jinan","Changchun","Kunming","Changsha","Taiyuan","Xiamen","Hefei","Shijiazhuang","Urumqi","Fuzhou","Wuxi","Zhongshan","Wenzhou","Nanning","Nanchang","Ningbo","Guiyang","Lanzhou","Zibo","Changzhou","Xuzhou","Tangshan","Baotou","Huizhou","Yantai","Shaoxing","Liuzhou","Nantong","Luoyang","Yangzhou"]

for city in top_cities:
    print(city)
    if city in rank_done:
        continue
    
    url = "https://api.waqi.info/feed/" + city + "/"
    qs = {"token":"8b556369620cbd18f0fc03b37e3959bd970f0706"}

    resp = rq.request("GET", url, params=qs)
    
    if resp.status_code != 200: 
        print("STATUS_CODE ERROR : ", resp.status_code, resp.text)
        continue
    print(city + " ok")
    json = resp.json()['data']
    
    if json == []:
        continue
    
    print(city + " ok 1")    
    try:
        co = json['iaqi']['co']['v']
        no2 = json['iaqi']['no2']['v']
        o3 = json['iaqi']['o3']['v']
        pm10 = json['iaqi']['pm10']['v']
        pm25 = json['iaqi']['pm25']['v']
        so2 = json['iaqi']['so2']['v']
        temp = json['iaqi']['t']['v'] 
    except KeyError:
        print()
    
    aqi = json['aqi']
    dom_pol = json['dominentpol']
    time = json['time']['s']
    
    ranking[1].append({
        "name": city,
        "aqi": aqi,
        "dom_pol": dom_pol,
        "co": co,
        "no2": no2,
        "o3": o3,
        "pm10": pm10,
        "pm25": pm25,
        "so2": so2,
        "temp": temp,
        "time": time
    })
    
    rank_done.append(city)
    
    with open('ranking_data.pkl', 'wb') as handle:
        pickle.dump(ranking, handle)
```
